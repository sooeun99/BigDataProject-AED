```python
# 1page : https://www.e-gen.or.kr/egen/search_aed.do?currentPageNum=1&jusoType=addrroad&lon=&lat=&searchType=general&sidoCode=11&emogdstr=&gugunCode=&addrroad=&loca=11&addrhosp=&dongCode=
# 2page : https://www.e-gen.or.kr/egen/search_aed.do?currentPageNum=2&jusoType=addrroad&lon=&lat=&searchType=general&sidoCode=11&emogdstr=&gugunCode=&addrroad=&loca=11&addrhosp=&dongCode=

# currentPageNum : 페이지 번호에 따라 URL이 변경됨을 확인(currentPageNum=1,2,3...)
```

* 구별 주소에서 중구 () -> 중구 

```python
# [CODE 564page까지만 확인 - 총 5641개 AED]
# url_base = 'https://www.e-gen.or.kr/egen/search_aed.do?currentPageNum='
# url_sub = '&jusoType=addrroad&lon=&lat=&searchType=general&sidoCode=11&emogdstr=&gugunCode=&addrroad=&loca=11&addrhosp=&dongCode='
k = 1

name = []
address = []
montime = []
tuetime = []
wedtime = []
thutime = []
fritime = []
sattime = []
suntime = []
holtime = []
while k!=565 :

    with urllib.request.ur lopen('https://www.e-gen.or.kr/egen/search_aed.do?currentPageNum='+ str(k) 
    + '&jusoType=addrroad&lon=&lat=&searchType=general&sidoCode=11&emogdstr=&gugunCode=&addrroad=&loca=11&addrhosp=&dongCode=') as response :
  
      html = (response.read())

    soup = BeautifulSoup(html, 'html.parser')

    # 설치기관명
    name.extend(m.text for m in soup.select('#placesList > li > div > h5 > a'))
    # 설치기관주소 데이터 추출 
    address.extend(a.text for a in soup.select('#placesList > li:nth-child(n) > div.info > p:nth-child(3)'))
    # #placesList > li:nth-child(1) > div > p:nth-child(3)
    
    # 사용시간 데이터 추출
    montime.extend(m.text for m in soup.select('#placesList > li:nth-child(n) > div > div.time-info > div:nth-child(2)'))
    tuetime.extend(tu.text for tu in soup.select('#placesList > li:nth-child(n) > div > div.time-info > div:nth-child(3)'))
    wedtime.extend(w.text for w in soup.select('#placesList > li:nth-child(n) > div > div.time-info > div:nth-child(4)'))
    thutime.extend(th.text for th in soup.select('#placesList > li:nth-child(n) > div > div.time-info > div:nth-child(5)'))
    fritime.extend(f.text for f in soup.select('#placesList > li:nth-child(n) > div > div.time-info > div:nth-child(6)'))
    sattime.extend(sa.text for sa in soup.select('#placesList > li:nth-child(n) > div > div.time-info > div:nth-child(7)'))
    suntime.extend(su.text for su in soup.select('#placesList > li:nth-child(n) > div > div.time-info > div:nth-child(8)'))
    holtime.extend(ho.text for ho in soup.select('#placesList > li:nth-child(n) > div > div.time-info > div:nth-child(9)'))
    
    k+=1
    
aed = pd.DataFrame({'설치기관명': name,
                    '설치기관주소': address,
                    '월요일' : montime,
                    '화요일' : tuetime,
                    '수요일' : wedtime,
                    '목요일' : thutime,
                    '금요일' : fritime,
                    '토요일' : sattime,
                    '일요일' : suntime,
                    '공휴일' : holtime,
                     })
    
aed['설치기관명'] = aed['설치기관명'].str.replace(pat=r'상세보기',repl=r'',regex=True)    
aed['월요일'] = aed['월요일'].str.replace(pat=r'[^0-9:~]+', repl= r'', regex=True)
aed['화요일'] = aed['화요일'].str.replace(pat=r'[^0-9:~]+', repl= r'', regex=True)
aed['수요일'] = aed['수요일'].str.replace(pat=r'[^0-9:~]+', repl= r'', regex=True)
aed['목요일'] = aed['목요일'].str.replace(pat=r'[^0-9:~]+', repl= r'', regex=True)
aed['금요일'] = aed['금요일'].str.replace(pat=r'[^0-9:~]+', repl= r'', regex=True)
aed['토요일'] = aed['토요일'].str.replace(pat=r'[^0-9:~]+', repl= r'', regex=True)
aed['일요일'] = aed['일요일'].str.replace(pat=r'1, 2, 3, 4, 5 째주', repl= r'', regex=True)
aed['일요일'] = aed['일요일'].str.replace(pat=r'[^0-9:~]+', repl= r'', regex=True)
aed['공휴일'] = aed['공휴일'].str.replace(pat=r'[^0-9:~]+', repl= r'', regex=True)

aed
```

```python
# 설치기관주소 구별설치기관주소(--구)
# 중구 정리 
addressgu = aed.설치기관주소.str.slice(start=6, stop=10)
addressgu = addressgu.str.replace('중구 난','중구')
addressgu = addressgu.str.replace('중구 남','중구')
addressgu = addressgu.str.replace('중구 다','중구')
addressgu = addressgu.str.replace('중구 덕','중구')
addressgu = addressgu.str.replace('중구 동','중구')
addressgu = addressgu.str.replace('중구 마','중구')
addressgu = addressgu.str.replace('중구 만','중구')
addressgu = addressgu.str.replace('중구 매','중구')
addressgu = addressgu.str.replace('중구 명','중구')
addressgu = addressgu.str.replace('중구 삼','중구')
addressgu = addressgu.str.replace('중구 새','중구')
addressgu = addressgu.str.replace('중구 서','중구')
addressgu = addressgu.str.replace('중구 세','중구')
addressgu = addressgu.str.replace('중구 소','중구')
addressgu = addressgu.str.replace('중구 손','중구')
addressgu = addressgu.str.replace('중구 수','중구')
addressgu = addressgu.str.replace('중구 을','중구')
addressgu = addressgu.str.replace('중구 왕','중구')
addressgu = addressgu.str.replace('중구 장','중구')
addressgu = addressgu.str.replace('중구 정','중구')
addressgu = addressgu.str.replace('중구 창','중구')
addressgu = addressgu.str.replace('중구 청','중구')
addressgu = addressgu.str.replace('중구 충','중구')
addressgu = addressgu.str.replace('중구 통','중구')
addressgu = addressgu.str.replace('중구 퇴','중구')
addressgu = addressgu.str.replace('중구 필','중구')
addressgu = addressgu.str.replace('중구 한','중구')
addressgu = addressgu.str.replace('중구 환','중구')
addressgu = addressgu.str.replace('중구 후','중구')
addressgu = addressgu.str.replace('별시 강','강동구')
addressgu = addressgu.str.replace('별시 노','노원구')
addressgu.head(50)
addressgu.tail(50)
```
