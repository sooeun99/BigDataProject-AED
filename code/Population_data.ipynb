```python
resident_population.drop(['동별(1)', '2018', '2018.2', '2018.3', '2018.4',
       '2018.5', '2019', '2019.2', '2019.3', '2019.4', '2019.5',
       '2020', '2020.2', '2020.3', '2020.4', '2020.5'], axis=1, inplace=True)
display(resident_population)
```

```python
resident_population.rename(columns = {'동별(2)':'소재지','2018.1':'2018','2019.1':'2019','2020.1':'2020'},inplace=True)
resident_population
```

```python
new_resident_population = resident_population.drop([0, 1], axis = 0)
new_resident_population
```

```python
new_resident_population = new_resident_population.reset_index()
display(new_resident_population)
new_resident_population.drop('index', axis=1, inplace=True)
new_resident_population
```

```python
new_resident_population['구별 거주인구 평균']=new_resident_population[['2018','2019','2020']].mean(axis=1)
```
