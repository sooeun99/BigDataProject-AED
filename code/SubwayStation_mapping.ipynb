```python
import pandas as pd
```

```python
inner_300m_count = pd.read_csv("C:/sooeun/DATAexam/전공프로젝트(빅데이터)/data/지하철역에서_가까운_AED_inner300m_개수.csv")
inner_300m_count
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome('C:/Temp/chromedriver.exe')
url = 'https://map.kakao.com/'
driver.get(url)
driver.implicitly_wait(time_to_wait=3) # 로딩될때까지 3초 wait

station_addr = []
for i in range(len(inner_300m_count)):
    print(i+1,'/',len(inner_300m_count), 'Processing..')
    search_box = driver.find_element(By.CSS_SELECTOR,'#search\.keyword\.query') # 검색 input 클릭
    search_box.clear() # 입력돼있던 텍스트 지우기
    search_box.send_keys(inner_300m_count['역이름'][i]) # 역 이름 하나씩 입력
    search_box.send_keys(Keys.RETURN) # Enter 입력
    time.sleep(1)
    addr = driver.find_element(By.CSS_SELECTOR,'div.info_item > div.addr > p:nth-child(1)') # 해당 역의 도로명주소 추출 
    station_addr.append(addr.text)
```

```python
station_addr = pd.DataFrame(station_addr)
station_addr.columns=['주소']
station_addr
# .str.contains('서울')
```

```python
inner_300m_count['주소'] = station_addr
inner_300m_count
```

```python
inner_300m_count = inner_300m_count[inner_300m_count['주소'].str.contains('서울')]
inner_300m_count
```

```python
inner_300m_count.reset_index(drop=True, inplace=True)
inner_300m_count
```

```python
inner_300m_count['주소'].apply(lambda x:x.split()[1])
```

```python
inner_300m_count['소재지'] = inner_300m_count['주소'].apply(lambda x:x.split()[1])
inner_300m_count
```
