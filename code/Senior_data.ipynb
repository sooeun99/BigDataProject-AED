```python
new_senior['2018년노인인구비율'] = pd.to_numeric(new_senior['2018.1']) / pd.to_numeric(new_senior['2018'])
new_senior['2019년노인인구비율'] = pd.to_numeric(new_senior['2019.1']) / pd.to_numeric(new_senior['2019'])
new_senior['2020년노인인구비율'] = pd.to_numeric(new_senior['2020.1']) / pd.to_numeric(new_senior['2020'])
```

```python
new_senior
```

```python
new_senior['노인인구비율'] = (new_senior['2018년노인인구비율']+new_senior['2019년노인인구비율']+new_senior['2020년노인인구비율'])/3
```
